```python
import psycopg2
from psycopg2 import Error

connection = None
cursor = None
schema_name = "project_management"

try:
    connection = psycopg2.connect(
        user="cristcas",
        password="13579",
        host="127.0.0.1",
        port="5432",
        database="sql_my_beloved"
    )
    
    cursor = connection.cursor()
    
    # Создание схемы
    cursor.execute(f"CREATE SCHEMA IF NOT EXISTS {schema_name};")
    print(f"Схема '{schema_name}' успешно создана или уже существует.")
    
    # Установка пути поиска
    cursor.execute(f"SET search_path TO {schema_name};")
    
    # Создание таблиц
    create_tables_sql = """
    CREATE TABLE IF NOT EXISTS employees (
        employee_id SERIAL PRIMARY KEY,
        first_name VARCHAR(50) NOT NULL,
        last_name VARCHAR(50) NOT NULL,
        email VARCHAR(100) UNIQUE,
        role VARCHAR(50)
    );
    
    CREATE TABLE IF NOT EXISTS projects (
        project_id SERIAL PRIMARY KEY,
        project_name VARCHAR(100) NOT NULL,
        start_date DATE,
        end_date DATE,
        status VARCHAR(20)
    );
    
    CREATE TABLE IF NOT EXISTS tasks (
        task_id SERIAL PRIMARY KEY,
        project_id INT REFERENCES projects(project_id),
        title VARCHAR(150) NOT NULL,
        due_date DATE,
        priority INT
    );
    
    CREATE TABLE IF NOT EXISTS assignments (
        assignment_id SERIAL PRIMARY KEY,
        task_id INT REFERENCES tasks(task_id),
        employee_id INT REFERENCES employees(employee_id),
        assigned_date DATE DEFAULT CURRENT_DATE
    );
    """
    
    cursor.execute(create_tables_sql)
    print(f"Таблицы в схеме '{schema_name}' успешно созданы.")
    
    connection.commit()
    print("Изменения сохранены.")
    
except (Exception, Error) as error:
    print(f"Ошибка при создании схемы/таблиц: {error}")
    if connection:
        connection.rollback()
        
finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
        print("Соединение закрыто.")
```

```python
import psycopg2
from psycopg2 import Error

connection = None
cursor = None
schema_name = "project_management"

try:
    connection = psycopg2.connect(
        user="cristcas",
        password="13579",
        host="127.0.0.1",
        port="5432",
        database="sql_my_beloved"
    )
    
    cursor = connection.cursor()
    
    # Установка пути поиска
    cursor.execute(f"SET search_path TO {schema_name};")
    
    # JOIN запрос: список задач с названиями проектов и именами сотрудников
    join_query = """
    SELECT 
        t.title AS задача,
        p.project_name AS проект,
        e.first_name || ' ' || e.last_name AS исполнитель,
        a.assigned_date AS дата_назначения
    FROM tasks t
    JOIN projects p ON t.project_id = p.project_id
    JOIN assignments a ON t.task_id = a.task_id
    JOIN employees e ON a.employee_id = e.employee_id;
    """
    
    cursor.execute(join_query)
    
    print(f"\n--- Результаты JOIN запроса ---")
    
    # Получение названий столбцов
    colnames = [desc[0] for desc in cursor.description]
    print("Столбцы:", colnames)
    print()
    
    # Получение и вывод всех записей
    records = cursor.fetchall()
    if records:
        for row in records:
            print(row)
    else:
        print("Запрос не вернул записей.")
    
    print("---\n")
    
except (Exception, Error) as error:
    print(f"Ошибка при выполнении JOIN запроса: {error}")
    
finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
        print("Соединение закрыто.")
```

```python
import psycopg2
import csv

connection = None
cursor = None
schema_name = "project_management"

try:
    connection = psycopg2.connect(
        user="cristcas",
        password="13579",
        host="127.0.0.1",
        port="5432",
        database="sql_my_beloved"
    )
    
    cursor = connection.cursor()
    cursor.execute(f"SET search_path TO {schema_name};")
    
    # JOIN запрос
    join_query = """
    SELECT 
        t.title AS задача,
        p.project_name AS проект,
        e.first_name || ' ' || e.last_name AS исполнитель,
        a.assigned_date AS дата_назначения
    FROM tasks t
    JOIN projects p ON t.project_id = p.project_id
    JOIN assignments a ON t.task_id = a.task_id
    JOIN employees e ON a.employee_id = e.employee_id;
    """
    
    cursor.execute(join_query)
    
    # Получение данных
    records = cursor.fetchall()
    colnames = [desc[0] for desc in cursor.description]
    
    # Сохранение в CSV
    with open('join_result.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(colnames)
        writer.writerows(records)
    
    print(f"CSV файл сохранён: join_result.csv")
    print(f"Сохранено записей: {len(records)}")
    
except (Exception, Error) as error:
    print(f"Ошибка: {error}")
    
finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
        print("Соединение закрыто.")
```

```python
import psycopg2
import sys

connection = None
cursor = None
schema_name = "project_management"

try:
    connection = psycopg2.connect(
        user="cristcas",
        password="13579",
        host="127.0.0.1",
        port="5432",
        database="sql_my_beloved"
    )
    
    cursor = connection.cursor()
    
    # Получаем все CREATE TABLE и INSERT для схемы
    cursor.execute(f"SET search_path TO {schema_name};")
    
    # Получаем структуру таблиц
    cursor.execute("""
        SELECT 
            'CREATE TABLE IF NOT EXISTS ' || table_name || ' (' ||
            string_agg(column_name || ' ' || data_type || 
                CASE WHEN is_nullable = 'NO' THEN ' NOT NULL' ELSE '' END, ', ') 
            || ');'
        FROM information_schema.columns 
        WHERE table_schema = 'project_management'
        GROUP BY table_name;
    """)
    
    create_statements = cursor.fetchall()
    
    # Получаем данные из таблиц
    tables = ['employees', 'projects', 'tasks', 'assignments']
    
    with open('project_management_dump.sql', 'w', encoding='utf-8') as f:
        f.write(f"-- Дамп схемы {schema_name}\n")
        f.write(f"CREATE SCHEMA IF NOT EXISTS {schema_name};\n\n")
        f.write(f"SET search_path TO {schema_name};\n\n")
        
        # Записываем CREATE TABLE
        for row in create_statements:
            f.write(row[0] + "\n\n")
        
        # Записываем данные из каждой таблицы
        for table in tables:
            cursor.execute(f"SELECT * FROM {table};")
            rows = cursor.fetchall()
            if rows:
                colnames = [desc[0] for desc in cursor.description]
                for row in rows:
                    values = []
                    for val in row:
                        if val is None:
                            values.append('NULL')
                        elif isinstance(val, str):
                            values.append(f"'{val}'")
                        elif isinstance(val, (int, float)):
                            values.append(str(val))
                        else:
                            values.append(f"'{val}'")
                    
                    insert = f"INSERT INTO {table} ({', '.join(colnames)}) VALUES ({', '.join(values)});"
                    f.write(insert + "\n")
                f.write("\n")
    
    print("Дамп успешно создан: project_management_dump.sql")
    
except (Exception, Error) as error:
    print(f"Ошибка: {error}")
    
finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
        print("Соединение закрыто.")
```
